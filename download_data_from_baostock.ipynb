## 从Python Baostock库下载sh.000300、sh.600519的股票数据，并存入CSV

```python
# 登陆系统
lg = bs.login()
# 显示登陆返回信息
print('login respond error_code:'+lg.error_code)
print('login respond  error_msg:'+lg.error_msg)
```

```python
rs = bs.query_history_k_data_plus("sh.000300",
    "date,code,open,high,low,close,preclose,volume,amount,pctChg",
    start_date='2019-5-01', end_date='2021-05-23', frequency="d")
print('query_history_k_data_plus respond error_code:'+rs.error_code)
print('query_history_k_data_plus respond  error_msg:'+rs.error_msg)
```

```python
# 打印结果集
data_list = []
while (rs.error_code == '0') & rs.next():
    # 获取一条记录，将记录合并在一起
    data_list.append(rs.get_row_data())
result = pd.DataFrame(data_list, columns=rs.fields)
# 结果集输出到csv文件
result.to_csv("000300.csv", index=False)
print(result)
```

## 读取两个表格的数据，进行计算收益率，并合并两表，再合入无风险收益率，为CAPM建模做准备

```python
data_000300=pd.read_csv('000300.csv',index_col=['date'])
```

```python
data_000300
```

```python
data_600519=pd.read_csv('600519.csv',index_col=['date'])
```

```python
data_600519
```

```python
#对000300计算收益率
data_ret_000300=np.log(data_000300['close']/data_000300['close'].shift(1)).cumsum()
```

```python
#将收益率作为一列
data_000300['ret_000300']=np.exp(data_ret_000300)
```

```python
data_000300
```

```python
#对600519计算收益率
data_ret_600519=np.log(data_600519['close']/data_600519['close'].shift(1)).cumsum()
```

```python
#将收益率作为一列
data_600519['ret_600519']=np.exp(data_ret_600519)
```

```python
data_600519
```

```python
#将两个表合并
data_con=pd.DataFrame(data_000300[['close','ret_000300']])
data_temp=pd.DataFrame(data_600519[['close','ret_600519']])
```

```python
data_con
```

```python
data_temp
```

```python
data_con=data_con.join(data_temp,lsuffix='_000300', rsuffix='_600519')
```

```python
data_con
```

```python
data_con=data_con.dropna()
data_con
```
